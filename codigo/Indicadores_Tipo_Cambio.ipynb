```python
# Importación de librerías regulares
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
```

```python
serie = "PN01207PM"
estructura = f"https://estadisticas.bcrp.gob.pe/estadisticas/series/api/{serie}/json"
data = requests.get(estructura).json()
```

```python
##  convertir a dataframe
df = pd.DataFrame(data["periods"])
df["values"] = df["values"].apply(lambda x: x[0])
df
```

```python
##implementar los indicadores de calidad

def calcular_metricas_calidad(df):
    # completitud
    completitud = df["values"].count() / df["values"].size
    print(f"Completitud: {completitud:.2%}")

    #Porcentaje de repetidos
    unicos = df["values"].duplicated().sum()/df["values"].size
    print(f"Porcentaje de repetidos: {unicos:.2%}")

    #porcentaje de nulos
    porcentaje_nulos = df["values"].isna().sum() / df["values"].size
    print(f"Porcentaje de nulos: {porcentaje_nulos:.2%}")

    metricas = {
        "completitud": completitud,
        "porcentaje_repetidos": unicos,
        "porcentaje_nulos": porcentaje_nulos
    }
    return metricas
```

```python
calcular_metricas_calidad(df)
```
